```python
import pandas as pd
```

# Load Dataset

```python
covid = pd.read_csv('/content/covid_19_indonesia_time_series_all.csv')
covid
```

```python
covid.info()
```

# Cleaning Data

Drop Duplicate

```python
covid = covid.drop_duplicates()
covid
```

Drop 'City or Regency' and 'Special Status' Columns

```python
covid = covid.drop(['City or Regency', 'Special Status'], axis=1)
covid
```

Turn "Date" Column into Date Data Type

```python
covid['Date'] = pd.to_datetime(covid['Date'])
covid['Date'].info()
```

Change 'Case Fatality Rate' and 'Case Recovered Rate' into numeric and remove the "%"

```python
# Check the column
covid[['Case Fatality Rate', 'Case Recovered Rate']]
```

```python
# Remove the "%" and turned it into numeric value
col_rate = ['Case Fatality Rate', 'Case Recovered Rate']
for i in col_rate:
  covid[i] = pd.to_numeric(covid[i].map(lambda x: x.rstrip('%')))
covid
```

Handle Outlier Values for 'Case Recovered Rate' and 'Case Fatality Rate'

```python
# Check how much 'Case Recovered Rate' that exceed 100%
covid['Case Recovered Rate'][covid['Case Recovered Rate'] > 100].count()
```

```python
covid.loc[(covid['Case Recovered Rate'] > 100), 'Case Recovered Rate'] = 100
covid['Case Recovered Rate']
```

```python
# Check how much 'Case Fatality Rate' that exceed 100%
covid['Case Fatality Rate'][covid['Case Fatality Rate'] > 100].count()
```

```python
covid.loc[(covid['Case Recovered Rate'] > 100), 'Case Recovered Rate'] = 100
covid['Case Recovered Rate']
```

```python
covid[['Case Fatality Rate', 'Case Recovered Rate']]
```

Change Every Null Value

```python
# Change all NaN values in 'Time Zone' into UTC+07:00
covid['Time Zone'] = covid['Time Zone'].fillna('UTC+07:00')
covid['Time Zone'].info()
```

```python
# Change all NaN values in 'Province' and 'Island' into Indonesia
covid['Province'] = covid['Province'].fillna('Indonesia')
covid['Island'] = covid['Island'].fillna('Indonesia')
covid[['Province','Island']].info()
```

```python
# Change all NaN in 'Total Cities', 'Growth Factor of New Cases', and 'Growth Factor of New Deaths' with mean of each their own column
col_mean = ['Total Cities','Growth Factor of New Cases','Growth Factor of New Deaths']
for i in col_mean:
  covid[i] = covid[i].fillna(covid[i].mean())
covid[col_mean].info()
```

```python
# Change all NaN in 'Total Urban Villages' and 'Total Rural Villages' with 0
covid['Total Urban Villages'] = covid['Total Urban Villages'].fillna(0)
covid['Total Rural Villages'] = covid['Total Rural Villages'].fillna(0)
covid[['Total Urban Villages','Total Rural Villages']].info()
```
